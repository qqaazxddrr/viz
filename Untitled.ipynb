```python
cities_coordinate={}
for i in range(len(cities)):
    for j in range(len(cities[i]['children'])):
        cities_coordinate[cities[i]['children'][j]['name']]=(float(cities[i]['children'][j]['log']),float(cities[i]['children'][j]['lat']))
```

```python
from math import radians, cos, sin, asin, sqrt,atan2
#计算两点间距离-m
def geodistance(lng1,lat1,lng2,lat2):
    lng1, lat1, lng2, lat2 = map(radians, [lng1, lat1, lng2, lat2])
    dlon=lng2-lng1
    dlat=lat2-lat1
    a=sin(dlat/2)**2 + cos(lat1) * cos(lat2) * sin(dlon/2)**2
    c=2 *atan2(sqrt(a), sqrt(1 - a))
    distance = 6373 * c
    return distance
```

```python
import networkx as nx
```

```python
G=nx.Graph()
```

```python
for key in cities_coordinate.keys():
    G.add_node(key)
```

```python
len(G.nodes())
```

```python
dis=geodistance(cities_coordinate['北京'][0],cities_coordinate['北京'][1],cities_coordinate['上海'][0],cities_coordinate['上海'][1])
```

```python
dis
```

```python
for city1 in cities_coordinate.keys():
    for city2 in cities_coordinate.keys():
        dis=geodistance(cities_coordinate[city1][0],cities_coordinate[city1][1],cities_coordinate[city2][0],cities_coordinate[city2][1])
        if (dis<300) and (city1 is not city2):
            G.add_edge(city1,city2)
```
